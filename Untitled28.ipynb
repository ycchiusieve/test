```python
import numpy as np
import pandas as pd
def solve(num, ck_grid, sample=False):
    result=[]
    ck=[]
    poslist=permutations(num)
    for p in poslist:
        rt_sum=None
        for grid in ck_grid:
            sums=sum(p[pt] for pt in grid)
            if rt_sum==None: rt_sum=sums
            elif rt_sum!=sums or (sample==True and sums in ck): break
        else: 
            result.append([p,rt_sum])
            ck.append(rt_sum)
    result=pd.DataFrame(sorted(result,key=lambda x: x[1]),columns=['solution','sum'])
    return result

```

```python
from math import factorial
from itertools import permutations
import pandas as pd
def solve2(space, L):
    lattices=L[:]   
    grid = []
    diff = []
    solution=[]
    skt = set()
    tmp=[0 for j in range(len(space))]
    for lattice in lattices:
        if not set(lattice) <= skt: 
            diff.append( list(set(lattice)-skt))
            grid.append(lattice)
            skt=skt.union(set(lattice))
               
        # print(skt)
    for g in grid: lattices.remove(g)
    
    def bruce(stage, re_sum):
        nonlocal grid
        nonlocal solution
        nonlocal tmp
        nonlocal space
        nonlocal lattices
        
        #print(stage)
        if stage == len(grid): 
            for lattice in lattices:
                if sum(tmp[i] for i in lattice)!=re_sum: return False        
            solution.append([tmp[:],re_sum])
            #print(solution)
            return True
        if stage == 0: 
            pt=0
            ct=0
            
            tmp=[None for j in range(len(space))]
            
        if stage<len(grid):
            ll=set(space)-set(x for x in tmp if x!= None)
            for t in permutations(ll,len(diff[stage])):
                
                if stage==0: 
                    re_sum=None
                    ct+=1
                    if int(100*ct/(factorial(len(space))/factorial(len(space)-len(diff[0]))))>=pt+5:
                             pt+=5
                             print(pt, end=' ')
                    #print(pt)
                tsum=sum(t)
              
                for k in range(len(diff[stage])):
                    tmp[diff[stage][k]]=t[k]
                for d in grid[stage]:
                    if d not in diff[stage]: tsum+=tmp[d]
                if re_sum==None: re_sum=tsum
                if tsum==re_sum:
                    bruce(stage+1,re_sum)
                for k in range(len(diff[stage])):
                    tmp[diff[stage][k]]=None
        
                
    print(grid, lattices, diff, sep='\n')
    bruce(0,0)
    result=pd.DataFrame(sorted(solution,key=lambda x: x[1]),columns=['solution','sum'])                
    
    return result
```

```python
import numpy as np
import pandas as pd
def solvewt(num, ck_grid,wt, sample=False):
    result=[]
    ck=[]
    poslist=permutations(num)
    for p in poslist:
        rt_sum=16
        for grid in ck_grid:
            sums=sum(p[pt] for pt in grid)+wt[ck_grid.index(grid)]
            if rt_sum==None: rt_sum=sums
            elif rt_sum!=sums or (sample==True and sums in ck): break
        else: 
            result.append([p,rt_sum])
            ck.append(rt_sum)
    result=pd.DataFrame(sorted(result,key=lambda x: x[1]),columns=['solution','sum'])
    return result
```

```python
for wt in permutations([0,1,2,3,4]):
    y=solvewt(range(1,10),[[0,1],[1,2,3],[3,4,5],[5,6,7],[7,8]],wt)
    if y.shape[0]>0: print(y)
```
